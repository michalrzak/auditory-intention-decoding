```python
shift = _get_shift_multiplier(40, 5000, 1000)
```

```python
shift_spectrum = np.fft.fftshift(np.fft.fft(shift[:, 0]))
```

```python
plt.plot(np.abs(shift_spectrum))
plt.show()
```
